#### Seasonal Decomposition

```python
import statsmodels.api as sm

#summer

#Create a monthly time series of bike ride (Average of all citibike stations)
df_new_sm=df_sm.mean(1).resample('M').mean()
df_new_sm.interpolate(inplace=True)

#Seasonal Decomposition using statsmodels api
res = sm.tsa.seasonal_decompose(df_new_sm.dropna())
plt.rcParams['figure.figsize']=(12,8)
resplot = res.plot()
```

```python
#winter

#Create a monthly time series of bike ride (Average of all citibike stations)
df_new_wt=df_wt.mean(1).resample('M').mean()
df_new_wt.interpolate(inplace=True)

#Seasonal Decomposition using statsmodels api
res = sm.tsa.seasonal_decompose(df_new_wt.dropna())
plt.rcParams['figure.figsize']=(12,8)
resplot = res.plot()
```

## Map CitiBike Data to CT level

### Load CitiBike station location data

```python
station_id=DataFrame.from_csv('Data/Station_id.csv')
#convert station_ids to string index
station_id.index=station_id.index.astype(str)
station_id.head()
```

### Create CitiBike station Geometry (Point) from Corrdinates (Lat & Lon)

```python
import shapely
# combine lat and lon to one column
station_id['lonlat']=zip(station_id.lon,station_id.lat)
# Create Point Geometry for based on lonlat column
station_id['geometry']=station_id[['lonlat']].applymap(lambda x:shapely.geometry.Point(x))
station_id.head()
```

### Intersect the two geometries

```python
a=[]
# check for each citibike stations to see what CT they are belong to
for i in range(len(station_id)):
    a.append(ct_shape[ct_shape.geometry.intersects(station_id.iloc[i].geometry)].BoroCT2010.values[0])
    print '\r',"%",str((i+1)*100./len(station_id))[:4],
station_id['BoroCT2010']=a
```

```python
station_id.BoroCT2010.head()
```

### Add Monthly rides to the station GeoDataFrame with CT code

```python
### Create average monthly rides column from monthly citibike data and add it to the station id which has the geometry
# for summer
station_id['MonthlyAve']=df_sm.sum()
# for winter
station_id_wt = station_id.copy()
station_id_wt['MonthlyAve']=df_wt.sum()
```

```python
#summer
station_id.head()
```

```python
#winter
station_id_wt.head()
```

### Add the monthly average rides to the CT shapefile 

```python
# summer
temp=station_id[['MonthlyAve']]
temp['station_id']=station_id.index
temp.index=station_id['BoroCT2010']
ct_shape_MonthlyAve=ct_shape.join(temp)
ct_shape_MonthlyAve.fillna(0,inplace=True)
```

```python
# winter
temp2=station_id_wt[['MonthlyAve']]
temp2['station_id']=station_id_wt.index
temp2.index=station_id_wt['BoroCT2010']
ct_shape_MonthlyAve2=ct_shape.join(temp2)
ct_shape_MonthlyAve2.fillna(0,inplace=True)
```

```python
# plot summer
f, ax = plt.subplots(figsize=(10,10))
ct_shape_MonthlyAve.plot(column='MonthlyAve',colormap='hot_r',alpha=1,linewidth=0.1,ax=ax)
plt.title("Monthly Ride Counts")
```

```python
# plot winter
f, ax = plt.subplots(figsize=(10,10))
ct_shape_MonthlyAve2.plot(column='MonthlyAve',colormap='hot_r',alpha=1,linewidth=0.1,ax=ax)
plt.title("Monthly Ride Counts")
```

```python
#winter
ct_shape_MonthlyAve_Zoom2=(ct_shape_MonthlyAve2[ct_shape_MonthlyAve
                           .centroid
                           .apply(lambda c:c.x>-74.1 and c.x<-73.92 and c.y>40.65 and c.y<40.8)])


print "For winter,"
print "Total number of CTs:", len(ct_shape)
print "Number of zoomed CTs:", len(temp2.MonthlyAve)
```

```python
import pysal as ps
```

### Spatial Weights

```python
qW_CT=ps.queen_from_shapefile(path)
```

#### Change the matrix to be row-standardized

```python
qW_CT.transform = 'r'
```

### Spatial Lag  
Similarity of CT neibors can be meseard by means of spatial lag as follow:
$$BikeRideCountLag = \sum_j w_{i,j} BikeRideCount$$

```python
# summer 
MonthlyRide=ct_shape_MonthlyAve_Zoom.MonthlyAve
MonthlyRideLag=ps.lag_spatial(qW_CT, MonthlyRide)

# winter
MonthlyRide2=ct_shape_MonthlyAve_Zoom2.MonthlyAve
MonthlyRideLag2=ps.lag_spatial(qW_CT, MonthlyRide2)
```

#### Create Quantiles for maping

```python
# summer
MonthlyRideLagQ10 = ps.Quantiles(MonthlyRideLag, k=10)
MonthlyRideLagQ10
```

```python
# winter
MonthlyRideLagQ10_2 = ps.Quantiles(MonthlyRideLag2, k=10)
MonthlyRideLagQ10_2
```

#### Map the monthly rides lag

```python
# summer
f, ax = plt.subplots(figsize=(9, 9))
ct_shape_MonthlyAve_Zoom.assign(cl=MonthlyRideLagQ10.yb).plot(column='cl', categorical=True, 
        k=10, cmap='hot_r', linewidth=0.1, ax=ax, 
        edgecolor='white', legend=True)
ax.set_axis_off()
plt.title("Summer Monthly Ride Spatial Lag Deciles")
```

```python
# winter
f, ax = plt.subplots(figsize=(9, 9))
ct_shape_MonthlyAve_Zoom2.assign(cl=MonthlyRideLagQ10_2.yb).plot(column='cl', categorical=True, 
        k=10, cmap='hot_r', linewidth=0.1, ax=ax, 
        edgecolor='white', legend=True)
ax.set_axis_off()
plt.title("Winter Monthly Ride Spatial Lag Deciles")
```

## Global  Spatial Autocorrelation

```python
# summer
b,a = np.polyfit(MonthlyRide, MonthlyRideLag, 1)
b,a 
```

```python
# winter
d,c = np.polyfit(MonthlyRide2, MonthlyRideLag2, 1)
d,c 
```

### Moran Scatterplot

```python
# summer
f, ax = plt.subplots(1, figsize=(9, 9))

plt.plot(MonthlyRide, MonthlyRideLag, '.', color='firebrick')

 # dashed vert at mean of the last year's PCI
plt.vlines(MonthlyRide.mean(), MonthlyRideLag.min(), MonthlyRideLag.max(), linestyle='--')
 # dashed horizontal at mean of lagged PCI
plt.hlines(MonthlyRideLag.mean(), MonthlyRide.min(), MonthlyRide.max(), linestyle='--')

# red line of best fit using global I as slope
plt.plot(MonthlyRide, a + b*MonthlyRide, 'r')
plt.title('Summer Moran Scatterplot')
plt.ylabel('Spatial Lag of MonthlyRides')
plt.xlabel('MonthlyRides')
```

```python
# winter
f, ax = plt.subplots(1, figsize=(9, 9))

plt.plot(MonthlyRide2, MonthlyRideLag2, '.', color='firebrick')

 # dashed vert at mean of the last year's PCI
plt.vlines(MonthlyRide2.mean(), MonthlyRideLag2.min(), MonthlyRideLag2.max(), linestyle='--')
 # dashed horizontal at mean of lagged PCI
plt.hlines(MonthlyRideLag2.mean(), MonthlyRide2.min(), MonthlyRide2.max(), linestyle='--')

# red line of best fit using global I as slope
plt.plot(MonthlyRide2, c + d*MonthlyRide2, 'r')
plt.title('Winter Moran Scatterplot')
plt.ylabel('Spatial Lag of MonthlyRides')
plt.xlabel('MonthlyRides')
```

### Moran's I

```python
# summer 
I_MonthlyRide = ps.Moran(MonthlyRide, qW_CT)
I_MonthlyRide.I, I_MonthlyRide.p_sim
```

```python
# winter
I_MonthlyRide2 = ps.Moran(MonthlyRide2, qW_CT)
I_MonthlyRide2.I, I_MonthlyRide2.p_sim
```

Thus, the $I$ statistic is $0.656$ for this data, and has a very small $p$ value.

### Hot and Cold spots 

```python
# summer
sig = LMo_MonthlyRide.p_sim < 0.01
hotspots = LMo_MonthlyRide.q==1 * sig
coldspots = LMo_MonthlyRide.q==3 * sig
print "for summer:"
print "Number of Hotspots:", hotspots.sum()
print "Number of Coldspots:", coldspots.sum()
```

```python
# winter
sig2 = LMo_MonthlyRide2.p_sim < 0.01
hotspots2 = LMo_MonthlyRide2.q==1 * sig
coldspots2 = LMo_MonthlyRide2.q==3 * sig
print "for winter:"
print "Number of Hotspots:", hotspots2.sum()
print "Number of Coldspots:", coldspots2.sum()
```

```python
from matplotlib import colors

# summer plotting
hcmap = colors.ListedColormap(['grey','red','blue'])
hotcold = hotspots*1 + coldspots*2

f, ax = plt.subplots(figsize=(9, 9))
ct_shape_MonthlyAve_Zoom.assign(cl=hotcold).plot(column='cl', categorical = True, k = 10, cmap = hcmap,
        linewidth=0.1, ax=ax, edgecolor='white', legend=True)
ax.set_axis_off()
plt.title("Summer MonthlyAve Spatial Lag Deciles")
```

```python
# winter plotting
hcmap2 = colors.ListedColormap(['grey','red','blue'])
hotcold2 = hotspots2*1 + coldspots2*2

f, ax = plt.subplots(figsize=(9, 9))
ct_shape_MonthlyAve_Zoom2.assign(cl=hotcold2).plot(column='cl', categorical = True, k = 10, cmap = hcmap2,
        linewidth=0.1, ax=ax, edgecolor='white', legend=True)
ax.set_axis_off()
plt.title("Winter MonthlyAve Spatial Lag Deciles")
```
