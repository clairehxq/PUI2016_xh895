# Normal Distribution: generate 100 random normal distributions with mean = 1

```python
#this cell 
import numpy as np
import matplotlib.pyplot as plt
import pylab as pl
%pylab inline
from scipy.stats import norm

#generate a list of 100 numbers suggesting sample sizes
sample_size = np.linspace(10,2000,100).astype(int)

#generate a dictionary normal_sample 
normal_sample = {}
normal_m = list()
normal_sd = list()

#a for loop to put len(normal_sample) normal distributions of different sizes in normal_sample dictionary.
#for each distribution, used numpy mean. 
#generated normal_sample['mean'] as a list containing 100 numbers suggesting the mean of each distribution
#generated normal_sample['std'] as a list containing 100 numbers suggesting the standard deviation of each distribution

my_mean = 1
my_std = 0.5

for i in range(0,100):
    normal_sample['%d'%i] = np.random.normal(my_mean, my_std, sample_size[i])
    normal_m.append(normal_sample['%d'%i].mean())
    normal_sd.append(normal_sample['%d'%i].std())

#plot    
normal_dist = pl.figure(figsize=(10,6)).add_subplot(111)
normal_dist.plot(sample_size, normal_m, 'go')
normal_dist.set_xlabel('sample size', fontsize=18)
normal_dist.set_ylabel('sample mean', fontsize=18)
normal_dist.set_title('Normal Distribution', fontsize=18)
normal_dist.plot([min(sample_size), max(sample_size)], [1, 1], 'k')    

#histogram
pl.figure(figsize=(10, 10))
pl.hist(normal_m,bins=30)
pl.xlabel('sample mean', fontsize = 18)
pl.ylabel('N', fontsize = 18)
pl.show()
```

## sample mean is more centered to the expected mean in larger sample size.
## error between sample mean and expected mean is smaller in larger sample size.

# Poisson Distribution: generate 100 random poisson distributions with mean = 1

# Binomial Distribution: generate random 100 bionmial distributions with mean = 1

```python
#generate a dictionary bi_sample
bi_sample={}
#declare bi_m and bi_sd are two lists.
bi_m = list()
bi_sd = list()

#a for loop to put len(normal_sample) binomial distributions of different sizes in bi_sample dictionary.
#for each distribution, used numpy mean and standard deviation. 
#generated bi_sample['mean'] as a list containing 100 numbers suggesting the mean of each distribution
#generated bi_sample['std'] as a list containing 100 numbers suggesting the standard deviation of each distribution

#random.binomial(n,p) takes two parameters, n suggesting the sample size and p the possibility of each single matter.
# np = my_mean, sqrt(np*(1-p)) = my_std


for i in range(0,100):
    p = 1 - (((my_std)**2)/my_mean)
    n = my_mean/p
    bi_sample['%d'%i] = np.random.binomial(n, p, sample_size[i])
    #normal_sample['mean'][i] = np.mean(normal_sample['%d'%i])
    #normal_sample['std'][i] = np.std(normal_sample['%d'%i])
    bi_m.append(bi_sample['%d'%i].mean())
    bi_sd.append(bi_sample['%d'%i].std())

#plot    
poisson_dist = pl.figure(figsize=(10,6)).add_subplot(111)
poisson_dist.plot(sample_size, bi_m, 'go')
poisson_dist.set_xlabel('sample size', fontsize=18)
poisson_dist.set_ylabel('sample mean', fontsize=18)
poisson_dist.set_title('Binomial Distribution', fontsize=18)
poisson_dist.plot([min(sample_size), max(sample_size)], [.75, .75], 'k')    
    
#histogram
pl.figure(figsize=(10, 10))
pl.hist(bi_m,bins=30)
pl.xlabel('sample mean', fontsize = 18)
pl.ylabel('N', fontsize = 18)
```

## sample mean is more centered to the expected mean in larger sample size.
## error between sample mean and expected mean is smaller in larger sample size.

# Chi-squared distribution: generate 100 random chi-squared distributions with mean = 1

```python
#generate a dictionary chi_sample
chi_sample={}
#declare chi_m and chi_sd are two lists.
chi_m = list()
chi_sd = list()

#a for loop to put len(chi_sample) chi-squared distributions of different sizes in chi_sample dictionary.
#for each distribution, used numpy mean and standard deviation. 
#generated chi_sample['mean'] as a list containing 100 numbers suggesting the mean of each distribution
#generated chi_sample['std'] as a list containing 100 numbers suggesting the standard deviation of each distribution

#random.chisquare(df,size) takes two parameters, df suggesting the degree of freedom , size suggesting sample size.


for i in range(0,100):
    chi_sample['%d'%i] = np.random.chisquare(my_mean, sample_size[i])
    #normal_sample['mean'][i] = np.mean(normal_sample['%d'%i])
    #normal_sample['std'][i] = np.std(normal_sample['%d'%i])
    chi_m.append(chi_sample['%d'%i].mean())
    chi_sd.append(chi_sample['%d'%i].std())

#plot    
poisson_dist = pl.figure(figsize=(10,6)).add_subplot(111)
poisson_dist.plot(sample_size, chi_m, 'go')
poisson_dist.set_xlabel('sample size', fontsize=18)
poisson_dist.set_ylabel('sample mean', fontsize=18)
poisson_dist.set_title('Chi-squared Distribution', fontsize=18)
poisson_dist.plot([min(sample_size), max(sample_size)], [1, 1], 'k')    

#histogram
pl.figure(figsize=(10, 10))
pl.hist(chi_m,bins=30)
pl.xlabel('sample mean', fontsize = 18)
pl.ylabel('N', fontsize = 18)
```

## sample mean is more centered to the expected mean in larger sample size.
## error between sample mean and expected mean is smaller in larger sample size.

# Cauchy distribution: generate 100 random standard Cauchy distributions(mean = 0)

```python
#generate a dictionary cauchy_sample
cauchy_sample={}
#declare cauchy_m and cauchy_sd are two lists.
cauchy_m = list()
cauchy_sd = list()

#a for loop to put len(cauchy_sample) standard cauchy distributions of different sizes in cauchy_sample dictionary.
#for each distribution, used numpy mean and standard deviation. 
#generated cauchy_sample['mean'] as a list containing 100 numbers suggesting the mean of each distribution
#generated cauchy_sample['std'] as a list containing 100 numbers suggesting the standard deviation of each distribution

#random.standard_cauchy(df,size) takes 1 parameters, size suggesting sample size. And set mean default to mean = 1


for i in range(0,100):
    cauchy_sample['%d'%i] = np.random.standard_cauchy(sample_size[i])
    #normal_sample['mean'][i] = np.mean(normal_sample['%d'%i])
    #normal_sample['std'][i] = np.std(normal_sample['%d'%i])
    cauchy_m.append(cauchy_sample['%d'%i].mean())
  

#plot    
poisson_dist = pl.figure(figsize=(10,6)).add_subplot(111)
poisson_dist.plot(sample_size, cauchy_m, 'go')
poisson_dist.set_xlabel('sample size', fontsize=18)
poisson_dist.set_ylabel('sample mean', fontsize=18)
poisson_dist.set_title('Standard Cauchy Distribution', fontsize=18)
poisson_dist.plot([min(sample_size), max(sample_size)], [1, 1], 'k')    
  
#histogram    
pl.figure(figsize=(10, 10))
pl.hist(cauchy_m,bins=30)
pl.xlabel('sample mean', fontsize = 18)
pl.ylabel('N', fontsize = 18)
```

## sample mean is more centered to the expected mean in larger sample size.
## error between sample mean and expected mean is smaller in larger sample size.

# Discovery:

1. The sample mean is more centered to the expected mean in larger sample size.

3. The average mean obtained from larger sample size is closer than that obtained from smaller sample size.
